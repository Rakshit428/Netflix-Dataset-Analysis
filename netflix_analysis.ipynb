```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
df.head()
```

```python
df['type'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
genre_split = df['listed_in'].str.split(', ', expand=True).stack()
genre_split = genre_split.reset_index(drop=True)
genre_split.value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top Countries by Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df.to_csv("netflix.csv", index=False)
```
